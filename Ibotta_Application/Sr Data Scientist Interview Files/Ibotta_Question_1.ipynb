# Ibotta Test

 Create a machine learning system which takes a seed audience of 2,323 Sprite purchasers
 and find a ‘similar’ set of users amongst the 27,055 provided customers. 
 The extended audience should be at least 2x the size of the seed audience.

You have input and output variable for existing customer (train data set) and only input variable for future customer (test data set). Now, you will build a model based on train data set and apply this
model on test data set to predict the behavior
Build different segment of customer based on input features only (known as segmentation) and identify the segments/ clusters of customers using unsupervised modeling techniques (K-means...). Hypothesis behind this technique is, all the customers of same clusters will behave similar compare to other clusters.
You can combine method 1 and 2 together for better prediction power also. Build clusters and build different model for each clusters and predict the behavior of customers.

# Feature Engineering

```python
import numpy as np
# Use customers list to create new column which identifies sprite customers
customers['sprite_customers'] = np.where(customers['id'].isin(sprite_cu_ids), 1, 0)
```

```python
# We look for columns which will provide quality information about customers
# Numerical Columns: ['age', 'number_of_children']
# Categorical Columns: ['gender', 'income', 'ethnicity']
# Columns get more data: ['zip']
customers.head()
```

```python
# Generate numerical data for income

income_df = pd.DataFrame({'income': customers.income.unique().tolist()})

lower = {'<25000': 0 , '25000-50000': 25000, '50000-75000': 50000, '75000-100000':75000, '>100000': 100000}
income_df['lower_bound_income'] = income_df['income'].map(lower)

higher = {'<25000': 25000 , '25000-50000': 50000, '50000-75000': 75000, '75000-100000':100000, '>100000': 125000}
income_df['higher_bound_income'] = income_df['income'].map(higher)

income_df
```

```python
# Generate numerical data for zip codes using lat long

lat_long_df = pd.read_csv("lat_long_data.csv", header=0 , names = ['zip','lat','long'] )
lat_long_df.head()
```

```python
# Merge the two pandas dataframes above, fill nulls
customers = customers.merge(income_df, how='left', on='income') \
    .merge(lat_long_df, how = 'left', on = 'zip').fillna(value='NaN')\
    .drop(labels = ['income', 'zip'], axis=1)\
    .set_index('id')
```

```python
customers.head()
```

```python
# One hot encode the categorical variables, and drop the categorical columns
one_hot_gender = pd.get_dummies(customers['gender'])
one_hot_gender.columns = ['gender_f', 'gender_m', 'gender_NaN']
one_hot_ethnicity = pd.get_dummies(customers['ethnicity'])
one_hot_ethnicity.columns = ["ethnicity_african_a", "ethnicity_asian","ethnicity_white", "ethnicity_hispanic", "ethnicity_NaN", "ethnicity_other"]
customers = customers.join(one_hot_gender).join(one_hot_ethnicity)
customers = customers.drop('gender', axis=1).drop('ethnicity', axis=1)
```

```python
# Get labels and customer ids as list

labels = customers['sprite_customers'].tolist()
customers = customers.drop('sprite_customers', axis=1)
```

```python
# Get customer ids as list
customer_ids = customers.index.values

#Send dataframe 
customer_ndarray = customers.as_matrix()
```

```python
# Impute missing values using the average

from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp.fit(customer_ndarray)
customer_ndarray = imp.transform(customer_ndarray)
```

```python
# Normalize using L1 or L2 regularization

from sklearn.preprocessing import normalize, StandardScaler


std_scale = StandardScaler().fit(customer_ndarray)
customer_ndarray = std_scale.transform(customer_ndarray)
```

# Check Variance Associated with Clusters, and Optimal Cluster Number

```python
from scipy.cluster.hierarchy import dendrogram, linkage
# generate the linkage matrix
Z = linkage(customer_ndarray , 'ward')
```

```python
last = Z[-10:, 2]
last_rev = last[::-1]
idxs = np.arange(1, len(last) + 1)
plt.plot(idxs, last_rev)

acceleration = np.diff(last, 2)  # 2nd derivative of the distances
acceleration_rev = acceleration[::-1]
plt.plot(idxs[:-2] + 1, acceleration_rev)
plt.show()
k = acceleration_rev.argmax() + 2  # if idx 0 is the max of this we want 2 clusters
print ("clusters:", k)
```

```python
plt.title('Hierarchical Clustering Dendrogram (truncated)')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=3,  # show only the last p merged clusters
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True  # to get a distribution impression in truncated branches
)
plt.show()
```

This dendrogram here shows variance as the height, and we can see that variance is mostly even across all three clusters which is a good sign. Although the optimal clusters chart above recommends that 9 clusters might be the most appropriate for this data set.
